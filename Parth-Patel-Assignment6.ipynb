```python
#Importing our usual packages
import bqplot
import numpy as np
import ipywidgets
import pandas as pd
```

```python
#Reading buildings Dataset, passing 0 for na values
buildings = pd.read_csv("https://uiuc-ischool-dataviz.github.io/is445_AOUAOG_fall2021/week02/data/building_inventory.csv"
                       ,na_values = {'Square Footage': 0,
                                    'Year Constructed':0,
                                    'Year Acquired':0})
```

```python
buildings.columns
```

```python
buildings
```

```python
#For the heat map, I will create a pivot table with index - Agency Name and Columns - County.
#Values will be the mean of Square footage for that set of criteria.
#To do that, I will group by Agency name and County, and find out mean square footage for those columns.
data = buildings[['County','Agency Name','Square Footage']].groupby(['Agency Name','County'])['Square Footage'].mean()
data
```

```python
#Converting the grouped data into a pivot table for the heat map. 
#There are a lot of NA values as for some combination of Agency Name and COunty, there arent any Square Footage valuee.
#fill_value parameter will convert those NA values to 0.
data = pd.pivot_table(buildings,index=['Agency Name'],columns=['County'],values=['Square Footage'],aggfunc='mean',fill_value=0)
data
```

```python
#Now to drive the histogram from the heatmap, I will need to pass the keys of the selected column.
#observe method will give the ith and jth values which I will use in the below syntax to fetch the keys.
print(data.index[0]) #This gives the agency name
print(data.columns[0][1]) #This gives the County name (Just data.columns[0][1] gives something like ('Square Footage', 'Adams')). So we need the 2nd value of this array which is the county name
```

```python
#plotting only the heatmap.

#Scales
col_sc = bqplot.ColorScale(scheme="RdPu")
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

#Defining our axes
ax_col = bqplot.ColorAxis(scale = col_sc,orientation = 'vertical', side = 'top')
ax_x = bqplot.Axis(scale = x_sc, label = 'Agencies')
ax_y = bqplot.Axis(scale = y_sc,orientation ='vertical', label = 'County')

#Defining mark for heatmap
heat_map = bqplot.GridHeatMap(color = data.values, scales = {'color':col_sc,'row':x_sc,'column':y_sc},
                            interactions = {'click':'select'},
                            anchor_style = {'fill':'blue'},
                            selected_style = {'opacity':1.0},
                            unselected_style = {'opacity':0.8})

#The final plot
fig = bqplot.Figure(marks = [heat_map],axes = [ax_col,ax_x,ax_y,])
fig
```
