So If for example user wants to locate the exact location, in for example Chicago city where the UFO was seen, he/she can visualize the exact location.
```python
#Lets get interactive!!
@ipywidgets.interact(color_map=plt.colormaps()) #Passed the list of colormap for user to choose from
def choose_cmap(color_map): #created a function to display the plot with the selected colormap passed as a parameter
    plt.scatter(ufo_final['latitude'], ufo_final['longitude'],c=ufo_final['duration_seconds'],cmap = color_map) 
#Passed the selected colormap to the 'cmap' parameter to get color map that user selected
    plt.colorbar()
    plt.show()
```

```python
@ipywidgets.interact(color_map=plt.colormaps()) 
def choose_cmap(color_map): 
    plt.scatter(ufo_final['latitude'], ufo_final['longitude'],c=np.log10(ufo_final['duration_seconds']),cmap = color_map) 
    plt.colorbar()
    plt.show()
```

Lets get started with the second widget!!!!

```python
#Plotting out a simple Bar graph with two fields (1 categorical and 1 numerical)
#Initially i tried out without creating the axis. But the plot size was small and labels were overlapping with each other.
#So, created subplots to plot the fields
fig, ax = plt.subplots(1,1,figsize=(20,10))
ax.bar(ufo_final['shape'],ufo_final['duration_seconds'])
plt.show()
```

```python
#making the above plot interactive!!
#Here duration_seconds will remain constant, and user will choose which field to bin from the dropdown.
#asking for  usier input - C for Categorical columns and N for Numerical columns
choose=input("Press C for categorical and N for Numerical")
if choose=='C': #If user selected categorical columns
    @ipywidgets.interact(choose_columns=['shape','state','country']) #List of categorical columns to choose from
    def plot_bar(choose_columns):
        fig, ax = plt.subplots(1,1,figsize=(20,10))
        ax.bar(ufo_final[choose_columns],ufo_final['duration_seconds'])
        plt.show()
else: #If user selected numerical Columns
    @ipywidgets.interact(choose_bins=range(10,200),cols=['longitude','latitude','duration_seconds']) #Passing numerical columns
#Providing a bin range from 10-200 for user to choose or change the bin number size according to what looks best!!
    def make_plot(cols,choose_bins):
        ufo_final.hist(column=cols,bins=int(choose_bins),figsize=(15,6))
        plt.show()
```

For the bar plot, I chose state, shape, country , latitude and longitide as they were feasible to display. Fields like city, dates and comments have large number of unique values, so aggregating them is not so feasible. Hence I chose only those values which have less unique values.
